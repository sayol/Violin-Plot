```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset(name="mpg")
cars
```

```python
cars.info()
```

```python
cars.describe()
```

```python
cars.isnull().sum()
```

```python
df = cars.dropna()
df
```

```python
sns.violinplot(data=df["displacement"])
plt.show()
```

```python
df["cylinders"].unique()
```

```python
sns.violinplot(data=df,x="cylinders",y="displacement",hue="origin")
plt.show()
```

```python
sns.boxplot(data=df, x="cylinders", y="displacement", hue="origin")
plt.show()
```

```python
sns.histplot(data=df, x="cylinders", y="displacement", hue="origin")
plt.show()
```

```python
sns.scatterplot(data=df, x="cylinders", y="displacement", hue="origin")
plt.show()
```

```python
sns.violinplot(data=df[df["origin"].isin(["japan", "europe"])],
               x="cylinders",
               y="displacement",
               hue="origin")
plt.show()
```

```python
sns.violinplot(data=df[df["origin"].isin(["japan", "europe"])],
               x="cylinders",
               y="displacement",
               hue="origin",
               split=True)
plt.show()
```

```python
sns.violinplot(data=df[df["origin"].isin(["japan", "europe"])],
               x="cylinders",
               y="displacement",
               hue="origin",
               scale_hue=False,
               split=True)
plt.show()
```
